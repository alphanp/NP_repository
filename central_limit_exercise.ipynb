```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
pop1 = np.random.binomial(10,0.2,10000)
print(pop1.mean(),pop1.std())
```

```python
pop2 = np.random.binomial(10,0.5,10000)
print(pop2.mean(),pop2.std())
```

```python
plt.hist(pop1)

plt.show()
```

```python
plt.hist(pop2)
plt.show()
```

```python
sample1=np.random.choice(pop1,100,replace=True)
sample2=np.random.choice(pop2,100,replace=True)
```

```python
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
print(sample1.mean(),sample1.std())
print(sample2.mean(),sample2.std())
```

1. Increase the sample size from 100 to 1000 - expect mean and sd to get close to the population values of mean and SD

```python
sample3a=np.random.choice(pop1,1000,replace=True)
sample4a=np.random.choice(pop2,1000,replace=True)
plt.hist(sample3a, alpha=0.5, label='sample 1') 
plt.hist(sample4a, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print(sample3a.mean(),sample3a.std())
print(sample4a.mean(),sample4a.std())
```

```python
sample3b=np.random.choice(pop1,20,replace=True)
sample4b=np.random.choice(pop2,20,replace=True)
plt.hist(sample3b, alpha=0.5, label='sample 1') 
plt.hist(sample4b, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()

print(sample3b.mean(),sample3b.std())
print(sample4b.mean(),sample4b.std())
```

2.	Change the probability value (p in the NumPy documentation) for pop1 to 0.3, then take new samples and compute the 
t-statistic and p-value. Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why?

```python
pop3 = np.random.binomial(10,0.3,10000)
print(pop3.mean(),pop3.std())

sample_pop3 = np.random.choice(pop3,1000,replace=True)
plt.hist(sample_pop3, alpha=0.5, label='sample pop3')
plt.legend(loc='upper right') 
plt.show()
```

```python
print(ttest_ind(sample_pop3, sample3a, equal_var=False))
```

```python
pop4 = np.random.binomial(10,0.4,10000)
print(pop4.mean(),pop4.std())

sample_pop4 = np.random.choice(pop3,1000,replace=True)
plt.hist(sample_pop3, alpha=0.5, label='sample pop4')
plt.legend(loc='upper right') 
plt.show()
print(ttest_ind(sample_pop4, sample3a, equal_var=False))
```

3.	Change the distribution of your populations from binomial to a distribution of your choice. Do the sample mean values still
accurately represent the population values?

```python
rand1 = np.random.normal(4,1.55,10000)
plt.hist(rand1,bins=10)
plt.show()

sample1_rand1 = np.random.choice(rand1,1000,replace=True)
print(sample1_rand1.mean(),sample1_rand1.std())

sample2_rand1 = np.random.choice(rand1,100,replace=True)
print(sample2_rand1.mean(),sample2_rand1.std())
```

As seen above the sample mean and SD are close to population values even when normal distribution is sampled.

```python
rand2 = np.random.normal(8,2.55,10000)
plt.hist(rand2,bins=20)
plt.show()

sample1_rand2 = np.random.choice(rand2,1000,replace=True)
print(sample1_rand2.mean(),sample1_rand2.std())

sample2_rand2 = np.random.choice(rand2,100,replace=True)
print(sample2_rand2.mean(),sample2_rand2.std())
```

As seen above the sample mean and SD are close to population values even when normal distribution is sampled.
